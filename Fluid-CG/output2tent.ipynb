```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.pyplot import cm
```

```python
import csv
results = []
with open('test', 'r') as csvfile:
    reader = csv.reader(csvfile, delimiter=',', quotechar='|')
    for row in reader:
        results.append(row)
```

```python
import numpy as np
cellNum = 128
X = [i for i in np.linspace(1,cellNum,cellNum)]
Y = [i for i in np.linspace(1,cellNum,cellNum)]
X, Y = np.meshgrid(X, Y)

Z = np.array(results)
Z = Z.astype(np.float)
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap = cm.coolwarm, linewidth = 0)
plt.show()
```
